**Membuat dataset dengan saya beri nama "data" yang berisi file stock.csv**

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("stock.csv")
```

**Distribusi data pada setiap atribut Purchase Price**



```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.hist(data['Purchase Price'], range=(1,12))
plt.title('Distribusi data pada setiap atribut Purchase Price',fontsize=15, color='red')
plt.xlabel('Purchase Price', fontsize = 12)
plt.ylabel('Jumlah Banyaknya', fontsize = 12)
plt.xlim(xmin=0,xmax=12)
plt.show()
```

**Distribusi data pada setiap atribut Current Price**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.hist(data['Current Price'], range=(1,12))
plt.title('Distribusi data pada setiap atribut Current Price',fontsize=15, color='red')
plt.xlabel('Current Price', fontsize = 12)
plt.ylabel('Jumlah Banyaknya', fontsize = 12)
plt.xlim(xmin=0,xmax=12)
plt.show()
```

**Distribusi data pada setiap atribut Gain/Loss**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.hist(data['Gain/Loss'], range=(1,12))
plt.title('Distribusi data pada setiap atribut Gain/Loss',fontsize=15, color='red')
plt.xlabel('Gain/Loss', fontsize = 12)
plt.ylabel('Jumlah Banyaknya', fontsize = 12)
plt.xlim(xmin=0,xmax=12)
plt.show()
```

**Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi**

```python
Dividend_share = (data.groupby('Stock Name')['Dividend/share']
                  .sum()
                  .reset_index()
                  .sort_values(by='Dividend/share', ascending=False)
                  .head(5))
print(Dividend_share)
```

**Data Visualization Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi**

```python
import matplotlib.pyplot as plt
Stock_Name=['Apple','Qualcomm','Microsoft','Cisco','Intel']
Dividend_share=['$2.28','$2.12','$1.56','$1.16','$1.09']
Shares=['100','200','150','250','500','100','200','225','185','50','100','1000','500']

plt.plot(Stock_Name,Dividend_share, label='Dividend share',marker='o',markerfacecolor='green')



plt.xlabel('Stock Name', fontsize = 20)
plt.ylabel('Dividend share', fontsize = 20)

plt.title('Stock Name yang Memiliki Dividen/Share 5 Tertinggi', fontsize = 20)

plt.legend(loc='upper center', bbox_to_anchor=(1.1, 1), shadow=True, ncol=1)

plt.gcf().set_size_inches(20, 6)

plt.show()
```

**Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi**

```python
Market_Value = (data.groupby('Stock Name')['Market Value']
                .sum()
                .reset_index()
                .sort_values(by='Market Value', ascending=False)
                .head(5))
print(Market_Value)
```

**Data Visualization Stock Name dan Symbol yang memiliki Market Value 5 Tertinggi**

```python
import matplotlib.pyplot as plt
Stock_Name=['Alphabet','Hawlett Packard','Redhat','Cisco','Twitter']
Market_Value=['$83.336.00','$8.842.50','8.626.00','$7.478.78','$7.302.55']
Shares=['100','200','150','250','500','100','200','225','185','50','100','1000','500']

plt.plot(Stock_Name,Market_Value, label='Market Value',color='yellow',marker='o',markerfacecolor='green')



plt.xlabel('Stock Name', fontsize = 20)
plt.ylabel('Market Value', fontsize = 20)

plt.title('Stock Name yang Memiliki Market Value 5 tertinggi', fontsize = 20)

plt.legend(loc='upper center', bbox_to_anchor=(1.1, 1), shadow=True, ncol=1)

plt.gcf().set_size_inches(20, 6)
plt.show()
```

**Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi**

```python
Annual_Yield = (data.groupby('Stock Name')['Annual Yield']
                .sum()
                .reset_index()
                .sort_values(by='Annual Yield', ascending=False)
                .head(7))
print(Annual_Yield)
```

**Data Visualization Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi**

```python
import matplotlib.pyplot as plt
Stock_Name=['Qualcomm','Cisco','Intel','Microsoft','Apple','Hewlett Packard Enterprise','Oracle']
Annual_Yield=['3.75%','3.49%','3.02%','2.38%','1.58%','1.47%','1.44%']
Shares=['100','200','150','250','500','100','200','225','185','50','100','1000','500']

plt.plot(Stock_Name,Annual_Yield, label='Annual Yield',color='purple',marker='o',markerfacecolor='black')



plt.xlabel('Stock Name', fontsize = 20)
plt.ylabel('Annual Yield', fontsize = 20)

plt.title('Stock Name yang Memiliki Annual Yield 7 tertinggi', fontsize = 20)

plt.legend(loc='upper center', bbox_to_anchor=(1.1, 1), shadow=True, ncol=1)

plt.gcf().set_size_inches(20, 6)
plt.show()
```
